```python
#ejercicio 2
import pandas as pd
def summary_notas(notas):
  notas = pd.Series(notas)
  resumen = pd.Series({
      'minima': notas.min(),
      'maxima': notas.max(),
      'media': notas.mean(),
      'desviacion tipica': notas.std()
  })
  return resumen

notas_alumnos = {'Juan': 8.5, 'María': 9.2, 'Pedro': 7.8, 'Ana': 9.5}
resumen = summary_notas(notas_alumnos)
print("\nResumen de notas:")
resumen
```

```python
#ejercicio 3
import pandas as pd
def passed_students(notas):
  notas = pd.Series(notas)
  passed = notas[notas >= 5]
  return passed.sort_values(ascending=False)

notas_alumnos = {'Juan': 8.5, 'María': 9.2, 'Pedro': 4.8, 'Ana': 9.5, 'Luis': 6.0}
aprobados = passed_students(notas_alumnos)
print("\nNotas de alumnos aprobados:")
aprobados
```

```python
#ejercicio 5
import pandas as pd
def calculate_monthly_balance(df, months):
    # Filtrar el DataFrame para incluir solo los meses especificados
    filtered_df = df[df['Mes'].isin(months)]

    # Calcular el balance para cada mes filtrado y luego sumar los balances
    total_balance = (filtered_df['Ventas'] - filtered_df['Gastos']).sum()

    return total_balance

# Ejemplo
selected_months = ['Febrero', 'Abril']
total_balance_selected_months = calculate_monthly_balance(df, selected_months)
print(f"\nBalance total para los meses {selected_months}: {total_balance_selected_months}")
```

```python
#ejercicio 6
import pandas as pd
def summarize_stock_data(file_path):
  # Leer el CSV en un DataFrame
  try:
    df = pd.read_csv(file_path)
  except FileNotFoundError:
    print(f"Error: El archivo '{file_path}' no fue encontrado.")
    return None

  # Seleccionar solo las columnas numéricas para el resumen
  numeric_cols = df.select_dtypes(include=['number']).columns

  # Calcular mínimo, máximo y media para cada columna numérica
  summary_df = pd.DataFrame({
      'Minimo': df[numeric_cols].min(),
      'Maximo': df[numeric_cols].max(),
      'Media': df[numeric_cols].mean()
  })

  return summary_df
```

```python
#ejercicio 8
import pandas as pd
import numpy as np
import datetime

# Generar un DataFrame con los datos de los cuatro ficheros.
files = ['emisiones-2016.csv', 'emisiones-2017.csv', 'emisiones-2018.csv', 'emisiones-2019.csv']
df_list = []
for file in files:
    try:
        df_list.append(pd.read_csv(file, sep=';'))
    except FileNotFoundError:
        print(f"Error: El archivo '{file}' no fue encontrado. Asegúrate de que los archivos estén presentes.")


df_emisiones = pd.concat(df_list, ignore_index=True)

# Filtrar las columnas del DataFrame
cols_to_keep = ['ESTACION', 'MAGNITUD', 'ANO', 'MES'] + [f'D{i:02d}' for i in range(1, 32)]
df_emisiones_filtered = df_emisiones[cols_to_keep]

# Reestructurar el DataFrame para que los valores de los contaminantes de los días aparezcan en una única columna.
df_emisiones_melted = df_emisiones_filtered.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'],
                                                var_name='DIA',
                                                value_name='VALOR')

# Convertir la columna DIA a un número entero
df_emisiones_melted['DIA'] = df_emisiones_melted['DIA'].str[1:].astype(int)

# Añadir una columna con la fecha
df_emisiones_melted['FECHA'] = pd.to_datetime(df_emisiones_melted[['ANO', 'MES', 'DIA']], errors='coerce')

# Eliminar las filas con fechas no válidas y ordenar el DataFrame
df_emisiones_cleaned = df_emisiones_melted.dropna(subset=['FECHA'])
df_emisiones_cleaned = df_emisiones_cleaned[~pd.isna(df_emisiones_cleaned['VALOR'])] # Eliminar filas con VALOR NaN
df_emisiones_cleaned = df_emisiones_cleaned.sort_values(by=['ESTACION', 'MAGNITUD', 'FECHA'])

# Mostrar por pantalla las estaciones y los contaminantes disponibles
print("Estaciones disponibles:")
print(df_emisiones_cleaned['ESTACION'].unique())
print("\nContaminantes disponibles:")
print(df_emisiones_cleaned['MAGNITUD'].unique())

# Crear una función que devuelva una serie con las emisiones para una estación, contaminante y rango de fechas
def get_emisiones_by_station_contaminant_date_range(df, estacion, magnitud, start_date, end_date):

    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    filtered_emisiones = df[(df['ESTACION'] == estacion) &
                            (df['MAGNITUD'] == magnitud) &
                            (df['FECHA'] >= start_date) &
                            (df['FECHA'] <= end_date)]
    return filtered_emisiones.set_index('FECHA')['VALOR']

# Mostrar un resumen descriptivo (mínimo, máximo, media, etc.) para cada contaminante.
print("\nResumen descriptivo por contaminante:")
print(df_emisiones_cleaned.groupby('MAGNITUD')['VALOR'].describe())

# Mostrar un resumen descriptivo para cada contaminante por distritos.
# Asumiendo que la columna 'ESTACION' se refiere a distritos o estaciones con códigos asociados a distritos.
# Si 'ESTACION' no representa distritos directamente, esta parte necesitaría un mapeo adicional.
print("\nResumen descriptivo por contaminante y estación:")
print(df_emisiones_cleaned.groupby(['MAGNITUD', 'ESTACION'])['VALOR'].describe())

# Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo
def get_summary_by_station_contaminant(df, estacion, magnitud):

    filtered_emisiones = df[(df['ESTACION'] == estacion) & (df['MAGNITUD'] == magnitud)]
    return filtered_emisiones['VALOR'].describe()

# Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados
def get_monthly_average_by_contaminant_year(df, magnitud, year):

    filtered_emisiones = df[(df['MAGNITUD'] == magnitud) & (df['ANO'] == year)].copy()
    filtered_emisiones['MES_STR'] = filtered_emisiones['MES'].astype(str).str.zfill(2)
    monthly_average = filtered_emisiones.groupby(['ESTACION', 'ANO', 'MES_STR'])['VALOR'].mean().reset_index()
    pivot_table = monthly_average.pivot_table(index='ESTACION', columns='MES_STR', values='VALOR')
    return pivot_table

# Crear un función que reciba una estación de medición y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.
def get_monthly_average_by_station(df, estacion):

    filtered_emisiones = df[df['ESTACION'] == estacion].copy()
    filtered_emisiones['MES_STR'] = filtered_emisiones['MES'].astype(str).str.zfill(2)
    monthly_average = filtered_emisiones.groupby(['MAGNITUD', 'ANO', 'MES_STR'])['VALOR'].mean().reset_index()
    pivot_table = monthly_average.pivot_table(index='MAGNITUD', columns=['ANO', 'MES_STR'], values='VALOR')
    return pivot_table
```
