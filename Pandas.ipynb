SERIES

```python
#crear una serie
serie2=pd.Series([2,5,7,4,9],index=['a','c','e','d','f'])
serie2
```

```python
#borrar series o filtros
serie2 = serie2.drop(['a', 'e'])
serie2
```

```python
#seleccionar filas
serie2[['c','d']]
```

```python
#asignacion
serie2['a']=100
serie2
```

```python
#operadores
serie3 = serie2 + serie2
serie3
```

```python
serie3 = serie2 - serie2
serie3
```

```python
serie3 = serie2 * serie2
serie3
```

```python
serie3 = serie2 / serie2
serie3
```
